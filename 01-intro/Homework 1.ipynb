```python
# Compute the duration in minutes
df_jan['duration'] = (df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
# Overview
df_jan.duration.head()
```

```python
# Calculate the standard deviation of trip durations in January
std_duration_january = df_jan['duration'].std()

print(f"The standard deviation of trip duration in January is {std_duration_january:.2f} minutes.")
```

## Question 3. Handling Ouliers in the duration columns


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot a boxplot for the duration column from jan_data
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_jan['duration'])
plt.title('Boxplot of Trip Durations')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
# Identify outliers based on a threshold (e.g., values greater than 60 and values less than 1)

# Durations between 1 and 60 minutes
jan_data = df_jan[(df_jan['duration'] >= 1) & (df_jan['duration'] <= 60)].reset_index(drop=True)
```

```python
jan_data.shape
```

```python
# Fraction of records left
fraction_left = jan_data.shape[0] / df_jan.shape[0]

# Print the fraction of records left
print(f"The fraction of records left after dropping outliers is: {fraction_left*100:.0f}%")
```

```python
# Plot the histogram for the filtered duration column
plt.figure(figsize=(10, 6))
sns.histplot(jan_data['duration'], bins=30, kde=True)
plt.title('Histogram of Trip Durations (1 to 60 minutes) in January')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
